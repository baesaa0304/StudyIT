```python
import numpy as np  # ndarray
import pandas as pd  # Series, DataFrame
import matplotlib.pyplot as plt  # 시각화(그래프)
import seaborn as sns  # 시각화, 예제 데이터
```

```python
tips = sns.load_dataset('tips')
```

#Boc plot

```python
# tip 컬럼의 box plot
sns.boxplot(data=tips, y='tip')
plt.show()
```

## 그룹(카테고리)별 box plot

```python
# 성별 tip의 box plot
sns.boxplot(data=tips, x='sex' , y ="tip")
plt.show()
```

```python
sns.boxplot(data=tips, x ='tip', y = 'sex')
plt.show()
```

요일별 tip의 box plot

```python
sns.boxplot(data=tips, x ='day', y='tip')
plt.show
```

* subplot 2개를 생성
* 각각의 subplot 흡연여부별 tips의 box plot, 시간별 box plot을 그리세요.

```python
fig, ax = plt.subplots(ncols=2 ,figsize=(8,4))

# 흡연 여부별 팁 상자 그림
sns.boxplot(data=tips, x="smoker", y="tip", ax=ax[0])

# 시간별 팁 상자 그림
sns.boxplot(data=tips, x="time", y="tip", ax=ax[1])

plt.show()
```

#막대 그래프

##`countplot()` 함수
카테고리 타입 변수의 빈도수를 막대로 표현한 그래프.


```python
# 성별 빈도수 세로 막대 그래프
sns.countplot(data=tips, x ="sex")
plt.show()
```

```python
# 성별 빈도수 가로 막대 그래프
sns.countplot(data=tips, y="sex")
```

```python
# 요일 빈도수 막대 그래프
sns.countplot(data=tips, x ="day")
plt.show()
```

```python
#요일별 성별 빈도수
tips.groupby(['day','sex']).count()
```

```python
sns.countplot(data=tips, x ='day' , hue ="sex")
# hue 파라미터 : 그룹(카레고리) 별로 다른 색깔을 설정할 때.
plt.show()
```

```python
sns.countplot(data=tips, x = 'sex', hue="day")

plt.show()
```

```python
tips.groupby(['sex', 'smoker']).tip.count()
```

```python
# 성별 , 흡연여부별 빈도수
sns.countplot(data=tips, y = "sex" , hue="smoker")

plt.show()
```

## `barplt()` 함수
* 그룹(카테고리) 변수의 별로 통계학(평균, 표준편차, 중앙값..)을 막대 그래프로 표현.
* 통계랑 계산 함수의 기본값은 평균(mean)
* 카테고리 타입 변수와 숫자 타입 변수를 함께 시각화.


```python
# 성별 tip의 평균
tips.groupby('sex').tip.mean()
```

```python
sns.barplot(data=tips , x = 'sex', y ='tip') #estimator = "mean" : 기본값 평균 계산
plt.show()
```

```python
tips.groupby('sex').tip.median() # 중위값 중앙값 , 50$
```

```python
#흡연 여부별 tip의 평균을 시각화.
sns.barplot(data = tips, x='smoker', y ='tip')
plt.show()
```

```python
tips.groupby(['sex' , 'smoker']).tip.mean()
```

```python
# 성별 흡연여부별 tip의 평균을 시각화
sns.barplot(data=tips, x='sex', y='tip', hue='smoker')
plt.show()
```

# 삼전도 ; `scatterplot()`

* tip ~ total_bill 삼전도

```python
sns.scatterplot(data=tips, x='total_bill', y='tip')

# 그래프 표시
plt.show()
```

* tip ~ total_bill
* 성별에 따라서 점의 색깔이 다르게 표현

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex')
plt.show()
```

* tip ~ total_bill
* 성별에 따라서 점의 색깔이 다르개 표현
* 흡연여부에 따라서 점의 모양을 다르게

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', style='smoker')

plt.show()
```

* tip ~ total_bill
* 성별에 따라서 점의 색깔이 다르개 표현
* 흡연여부에 따라서 점의 모양을 다르게
* 점의 크기를 인원수에 따라서 다르게 표현

```python
# scatter plot 그리기
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex", style="smoker", size="size")

plt.show()
```

# 카테고리 그래프 = `catplot()`
카테고리 별로 subplot들을 생성하고 통계값(평균, 중앙값, ...) subplot에 같은 종류의 그래프들을 시각화.


```python
#요일별 팁의 평균을 막대 그래프
sns.barplot(data=tips, x = 'day', y='tip', hue='smoker')

plt.show()
```

```python
sns.catplot(data=tips, x='day' , y='tip', col='smoker', kind='bar')
plt.show()
```

* 요일별 팀의 평균막대 그래프
* 시간(lunch, Dinner) 별로 subplot을 생성.

```python
sns.catplot(data=tips, x="day", y="tip", col="time", kind="bar")

plt.show()
```

* 요일별 팀의 평균 막대 그래프
* 흡연여부별로 subplot
* 시간별로 subplot

```python
sns.catplot(data=tips, x="day", y="tip", col="smoker", row="time", kind="bar")

plt.show()
```

* 요일별 성별 팀의 막대 그래프
* 시간별로 subplot을 그리기

```python
sns.catplot(data=tips, x = 'day', y ='tip', hue='sex', kind='bar', col ='time')
```

* 성별 팁의 box plot
* 요일별로 subplot

```python
sns.catplot(data=tips, x = 'sex' , y = 'tip', col='day', kind = 'box')
```
